# Quais são os números de registros nos ultimo anos?

```python
%matplotlib inline

import csv
import collections
from operator import itemgetter
from matplotlib import pyplot as plt
import pandas as pd
from operator import itemgetter
import rows



dados_lidos = rows.import_from_csv('registros_1996_a_2016.csv', encoding = "ISO-8859-1")
registros = rows.export_to_dicts(dados_lidos)
registros_validos = [ registro for registro in registros if registro['ano_do_obito'] != 'Total']

registros_descrentes = sorted(registros_validos, key=itemgetter('ano_do_obito'))
x = [registro['ano_do_obito'] for registro in registros_descrentes]
y = [int(registro['obitos_p_ocorrencia']) for registro in registros_descrentes]

serie = pd.Series(y, index=x)
serie.plot.bar(figsize=(20,8), colormap='BuPu_r', fontsize=25)
plt.xlabel('Anos')
plt.ylabel('Numero de registros')
plt.title("Numero registros de suicidios por ano", fontsize=25);

print(y[-2:])
```

# Qual o mês com o maior número de registros?

```python
%matplotlib inline

%matplotlib inline


import collections
import operator as op
from matplotlib import pyplot as plt
import pandas as pd

import rows


registros_por_mes = rows.import_from_csv('resgitro_por_meses_do_ano.csv', encoding = "ISO-8859-1")
registros_por_mes = rows.export_to_dicts(registros_por_mes)

registros_por_mes_ordenado = sorted(registros_por_mes[0].items(), key=op.itemgetter(1), reverse=True)
valores = [registro[1] for registro in registros_por_mes_ordenado]
meses = [registro[0].capitalize() for registro in registros_por_mes_ordenado]

serie = pd.DataFrame(valores, index=meses)
serie.plot.bar(subplots=True, figsize=(10, 5), fontsize=20)
plt.title("Registro por mes de 1996 à 2016", fontsize=20)
```

# Para qual sexo há mais registros ?

```python
%matplotlib inline

import csv
import collections
from operator import itemgetter
from matplotlib import pyplot as plt
import pandas as pd
from operator import itemgetter

import rows


# Ocorrencias por faixa de idade e sexo
dados_lidos = rows.import_from_csv('resgistro_por_sexo_por_ano.csv', encoding = "ISO-8859-1")
dados_por_sexo = rows.export_to_dicts(dados_lidos)
total_por_sexo = [registro for registro in dados_por_sexo if registro['ano_do_obito'] == 'Total']

por_sexo_valores = [total_por_sexo[0]['masc'], total_por_sexo[0]['fem']]
por_sexo_chave = ['Masculino', 'Feminino']

# Grafico por sexo
serie = pd.DataFrame(por_sexo_valores, index=por_sexo_chave)
serie.plot.pie(subplots=True, figsize=(10, 10), fontsize=20, autopct='%.2f%%')
plt.title("Registro por sexo de 1996 à 2016", fontsize=20)
```

# Para qual cor/raça há mais registros?

```python
%matplotlib inline


import collections
from operator import itemgetter
from matplotlib import pyplot as plt
import pandas as pd

import rows


registro_raca_2016 = rows.import_from_csv('registros_por_raca_cor.csv', encoding = "ISO-8859-1")
dados_por_raca = rows.export_to_dicts(registro_raca_2016)

valores = [valor for valor in dados_por_raca[0].values()]
racas = [chave.capitalize() for chave in dados_por_raca[0].keys()]

serie = pd.DataFrame(valores, index=racas)
serie.plot.pie(subplots=True, figsize=(9, 9), fontsize=20, autopct='%.2f%%')
plt.title("Registro por cor/raça de 1996 à 2016", fontsize=20)
```
