```python
%matplotlib inline

import csv
import collections
from matplotlib import pyplot as plt
import pandas as pd

# Lê do arquivo 
with open('registro_1996_a _2016.csv', encoding = "ISO-8859-1" ,newline='') as csvfile:
    registros = csv.DictReader(csvfile)
    for registro in registros:
        dados = registro

dados.pop('Total')
# Ordena por ano
dados_ordenados = collections.OrderedDict(sorted(dados.items()))

x = [ano for ano in dados_ordenados.keys()]
y = [int(valor) for valor in dados_ordenados.values()]

serie = pd.Series(y, index=x)
serie.plot.bar(figsize=(20,8), colormap='BuPu_r', fontsize=25)
plt.xlabel('Anos')
plt.ylabel('Numero de registros')
plt.title("Numero de suicidios por ano", fontsize=25);

print(y[-2:])
```

```python
%matplotlib inline

import csv
import collections
from operator import itemgetter
from matplotlib import pyplot as plt
import pandas as pd

import rows


# Ocorrencias por estado no ano de 2016
registro_2016 = rows.import_from_csv('A123225177_97_185_21.csv', encoding = "ISO-8859-1")
dados_2016 = rows.export_to_dicts(registro_2016)

rank = sorted(dados_2016, key=itemgetter('obitos_por_ocorrencia'), reverse=True)
del rank[0]
top_5 = rank[0:5]
estados = [estado['unidade_da_federacao'] for estado in top_5]
ocorrencias = [int(estado['obitos_por_ocorrencia']) for estado in top_5]

print(ocorrencias)
serie = pd.DataFrame(ocorrencias, index=estados)
serie.plot.bar(subplots=True, figsize=(8, 5), fontsize=22)
plt.title("Rank com mais registros em 2016", fontsize=22)
```

```python
%matplotlib inline

import csv
import collections
from operator import itemgetter
from matplotlib import pyplot as plt
import pandas as pd
from operator import itemgetter

import rows


# Ocorrencias por faixa de idade e sexo
registro_sexo_2016 = rows.import_from_csv('A200059189_4_74_34.csv', encoding = "ISO-8859-1")
dados_por_sexo_2016 = rows.export_to_dicts(registro_sexo_2016)


por_sexo_valores = [dados_por_sexo_2016[0]['obitos_por_ocorrencia'], 
                    dados_por_sexo_2016[1]['obitos_por_ocorrencia']]
por_sexo_chave = ['Masculino', 'Feminino']

# Grafico por sexo
serie = pd.DataFrame(por_sexo_valores, index=por_sexo_chave)
serie.plot.pie(subplots=True, figsize=(10, 10), fontsize=20, autopct='%.2f%%')
plt.title("Registro por sexo 2016", fontsize=20)
```

```python
%matplotlib inline


import collections
from operator import itemgetter
from matplotlib import pyplot as plt
import pandas as pd

import rows


# Ocorrencias por idade em 2016
registro_idade_2016 = rows.import_from_csv('A203417189_4_74_34.csv', encoding = "ISO-8859-1")
dados_por_idade = rows.export_to_dicts(registro_idade_2016)

dados_por_idade = [registro for registro in dados_por_idade if registro['obitos_por_ocorrencia'] is not None]

rank = sorted(dados_por_idade, key=itemgetter('obitos_por_ocorrencia'), reverse=True)

por_idade_valores = [dados['obitos_por_ocorrencia'] for dados in rank[1:-1]]
por_idade_chave = [dados['faixa_etaria'] for dados in rank[1:-1]]

# print(rank[1:-1])
# Grafico por sexo
serie = pd.DataFrame(por_idade_valores, index=por_idade_chave)
serie.plot.bar(subplots=True, figsize=(10, 5), fontsize=20)
plt.title("Registro por idade 2016", fontsize=20)
```
